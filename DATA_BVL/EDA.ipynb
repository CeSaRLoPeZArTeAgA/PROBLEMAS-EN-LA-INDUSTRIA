## Análisis de Portafolio de Inversion
---
Se analizara el modelo 

$$f(c_i,x_i)=c_1.X_1+\cdot +c_n.X_n$$

donde: 
n: numero de activos financieros
X_i: precio de la accion i

---

Consideremos un portafolio formado por $n$ activos financieros. 
Denotamos por

* $X_i \quad (i = 1,\dots,n)$ el precio (o valor) del activo $i$ en el horizonte temporal de interés
* $c_i \quad (i = 1,\dots,n)$ el peso invertida en dicho activo.

El valor total del portafolio puede escribirse como la función
\begin{equation*}
f(\mathbf{c}, \mathbf{X}) 
= \sum_{i=1}^{n} c_i X_i,
\end{equation*}

donde
$
\mathbf{c} = (c_1,\dots,c_n)^{\mathsf{T}}, 
\,\,
\mathbf{X} = (X_1,\dots,X_n)^{\mathsf{T}}.
$

En forma vectorial, el valor del portafolio se escribe de manera compacta como
\begin{equation*}
f(\mathbf{c}, \mathbf{X}) = \mathbf{c}^{\mathsf{T}} \mathbf{X}.
\end{equation*}


Los pesos del portafolio satisfacen:
\begin{equation*}
\sum_{i=1}^{n} c_i = 1,
\qquad
c_i \in \mathbb{R} \quad (i=1,\dots,n),
\end{equation*}
y
\begin{equation*}
c_i \ge 0 \quad \text{para todo } i.
\end{equation*}


---

Sea un portafolio compuesto por $n = 4$ activos financieros: ALICORP, CREDICORP, FERREYCORP y SOUTHERN. A partir de los precios diarios $X_{i,t}$, se definen los rendimientos simples como:
\begin{equation*}
    w_{i,t} = \frac{X_{i,t} - X_{i,t-1}}{X_{i,t-1}} = \frac{X_{i,t}}{X_{i,t-1}} - 1,
\end{equation*}
para cada activo $i = 1,2,3,4$ y cada día $t$.


Con estos rendimientos se construye la siguiente matriz:
\begin{equation*}
R = 
\begin{pmatrix}
w_{1,1} & w_{2,1} & w_{3,1} & w_{4,1} \\
w_{1,2} & w_{2,2} & w_{3,2} & w_{4,2} \\
\vdots & \vdots & \vdots & \vdots \\
w_{1,T} & w_{2,T} & w_{3,T} & w_{4,T}
\end{pmatrix}
\end{equation*}
donde cada columna corresponde a un activo y cada fila a un día.

La media de los rendimientos del activo $i$ se define como:
\begin{equation*}
    \bar{w}_i = \frac{1}{T} \sum_{t=1}^{T} w_{i,t}
\end{equation*}

```python
import pandas as pd
import numpy as np

df = pd.read_csv('DATA_UNIFICADA.csv')

# calcular rendimientos : (X_t - X_t-1) / X_t-1
rendimientos = df.copy()
rendimientos.set_index('Fecha', inplace=True)

#for col in rendimientos.columns:
#    rendimientos[col] = (rendimientos[col] - rendimientos[col].shift(1)) / rendimientos[col].shift(1)

for col in rendimientos.columns:
    rendimientos[col] =np.log( (rendimientos[col] ) / rendimientos[col].shift(1))

# eliminar primera fila (NaN porque no tiene dia anterior)
rendimientos = rendimientos.dropna()

rendimientos.head()
```

La matriz de covarianza $\Sigma$ del sistema de activos se define como:
\begin{equation*}
Cov=\Sigma =
\begin{pmatrix}
\text{Var}(w_1) & \text{Cov}(w_1,w_2) & \text{Cov}(w_1,w_3) & \text{Cov}(w_1,w_4) \\
\text{Cov}(w_2,w_1) & \text{Var}(w_2) & \text{Cov}(w_2,w_3) & \text{Cov}(w_2,w_4) \\
\text{Cov}(w_3,w_1) & \text{Cov}(w_3,w_2) & \text{Var}(w_3) & \text{Cov}(w_3,w_4) \\
\text{Cov}(w_4,w_1) & \text{Cov}(w_4,w_2) & \text{Cov}(w_4,w_3) & \text{Var}(w_4)
\end{pmatrix}
\end{equation*}



La covarianza entre los activos $i$ y $j$ mide cómo varían conjuntamente sus rendimientos:
\begin{equation*}
\text{Cov}(w_i, w_j) = \frac{1}{T - 1} \sum_{t=1}^{T} (w_{i,t} - \bar{w}_i)(w_{j,t} - \bar{w}_j)
\end{equation*}

```python
import pandas as pd
import numpy as np

# convertir a matriz numpy
R = rendimientos.values
T = len(R)  # numero de dias de datos
mean = np.mean(R, axis=0)  # promedio por activo

# matriz de covarianza manual 
cov = (R - mean).T @ (R - mean) / (T - 1)

# convertir a dataframe con nombres de columnas y filas
cov_df = pd.DataFrame(cov, 
        index=['ALICORP', 'CREDICORP', 'FERREYCORP', 'SOUTHERN'],
        columns=['ALICORP', 'CREDICORP', 'FERREYCORP', 'SOUTHERN'])

# impresion de la matriz de covarianza
cov_df
```

Pasamos a corroborar si la data es consistente a trabes de la martiz de covarianza $Cov$, para ello debemos
comprobar:
* Si es simetrica
* Si es semidefinida positiva
* Se vera los autovalores de la matriz

Si se comprueba que matriz Cov es semidefinida positiva y simetrica(PSD), se hara la
descompisicion de Cholesky directamente.

En caso que la matriz $Cov$ arroje que no es seminidefinida positiva, la factorización de Cholesky falla, para corregirlo, se añade una pequeña cantidad positiva que se suma a la diagonal de una matriz para hacerla definida positiva. La pequeña cantidad se le denomina jitter y se calcula como:

\begin{equation*}
\text{jitter} 
= 
\max(0, -\lambda_{\min})
+
\varepsilon \, \frac{\operatorname{tr}(Cov)}{n},
\qquad
\varepsilon = 10^{-12}.
\end{equation*}

Se define entonces la matriz ajustada:
\begin{equation*}
Cov_j = Cov + \text{jitter} \, I.
\end{equation*}

Esta nueva matriz $Cov_j$, se garantiza que es semidefinida positiva (SPD), por lo que:
\begin{equation*}
Cov_j = L L^{\mathsf{T}}.
\end{equation*}

Para comprobar la exactitud numérica, se calcula el error de reconstrucción:
\begin{equation*}
\| Cov_j - L L^{\mathsf{T}} \|_F,
\end{equation*}

donde $\|\cdot\|_F$ denota la norma de Frobenius.  


Idealmente:
\begin{equation*}
\| Cov_j - L L^{\mathsf{T}} \|_F \rightarrow 0.
\end{equation*}

```python
import numpy as np
import pandas as pd


# hacemos una copia, con lo cual trabajaremos
cov = cov_df.to_numpy(copy=True)

# comprobaciones de simetría y autovalores

# simetria numerica
is_symmetric = np.allclose(cov, cov.T, atol=1e-12, rtol=1e-12)

# autovalores (para matrices simetricas usa eigvalsh)
eigvals = np.linalg.eigvalsh(cov)

# tolerancia numerica dependiente de la escala de la matriz
scale = max(1.0, np.linalg.norm(cov, ord=2))
tol = 1e-12 * scale

# es semidefinida positiva y simetrica(PSD) si todos los autovalores >= -tol
is_psd = np.all(eigvals >= -tol)

print("######   Comprobacion de la matriz Cov  para descomposicion ######")
print("La matriz Cov es simetrica:", is_symmetric)
print("Los autovalores son:", eigvals)
print("La matriz Cov es semidefinida positiva (numéricamente):", is_psd)

# descomposicion Cholesky
#   - Si es SPD (lambda_min > tol), Cholesky directo
#   - Si no, añadimos el jitter mínimo necesario para volverla SPD
lam_min = eigvals.min()
n = cov.shape[0]

def cholesky_with_min_jitter(A, lam_min, tol, safety=1e-12):
    """
    Intenta Cholesky directo. Si falla o A no es SPD,
    añade jitter = max(0, -lam_min) + safety * tr(A)/n y reintenta.
    """
    try:
        L = np.linalg.cholesky(A)
        jitter_used = 0.0
        return L, jitter_used
    except np.linalg.LinAlgError:
        pass

    # jitter minimo para empujar el menor autovalor por encima de 0
    jitter = max(0.0, -lam_min) + safety * (np.trace(A)/A.shape[0] if A.shape[0] > 0 else 1.0)
    A_j = A + jitter * np.eye(A.shape[0])
    L = np.linalg.cholesky(A_j)
    return L, jitter

L, jitter = cholesky_with_min_jitter(cov, lam_min, tol)
print("")
print("######   Descomposicion Cholesky de la matriz Cov   ######")
print(f"Valor jitter añadido para descomposicion Cholesky: {jitter:.3e}")

# descomposicion Cholesky
print("Matriz L de descomposicion Cholesky")
L_df = pd.DataFrame(L, index=cov_df.index, columns=[f"c{i+1}" for i in range(n)])
print("\nFactor de Cholesky L (triangular inferior):")
print(L_df)

# verificacion 
recon = L @ L.T
res_norm = np.linalg.norm((cov + jitter*np.eye(n)) - recon, ord='fro')
print("")
print(f"Verificacio de descomposicion: ||A(+jitter) - L L^T||_F = {res_norm:.3e}")
```

## Definición de la Variable Aleatoria Normal Estándar

Se considera un vector aleatorio
$$
Z = (Z_1, Z_2, Z_3, Z_4)^\top
$$
cuyos componentes son variables aleatorias independientes e idénticamente distribuidas:
$$
Z_i \stackrel{\text{i.i.d.}}{\sim} \mathcal{N}(0,1), \qquad i=1,\dots,4
$$

Al simular \(n = 1000\) valores de cada componente, obtenemos una matriz:

$$
Z_{\text{mat}} \in \mathbb{R}^{1000 \times 4},
$$

donde cada fila representa una realización de \(Z\).


---

## Construcción de la Variable Aleatoria Multivariada $W$

Se define:

$$
W = L Z
$$

y para cada una de las 1000 simulaciones:

$$
W^{(k)} = L Z^{(k)},
\qquad k=1,\dots,1000.
$$

De forma vectorizada:

$$
W = Z L
$$

Por teoría, se garantiza que:

$$
W \sim \mathcal{N}(0, \widetilde{\text{Cov}}).
$$

---

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# genera 4 variables aleatorias Z1,...,Z4 iid ~  N(0,1)
n = 1000 # cada una con 1000 valores
Z = [np.random.randn(n) for _ in range(4)]

# la matriz Z_mat esta ya ordenado para la multiplicacion
Z_mat = np.column_stack(Z)    # matriz 1000 × 5

# genera X = L Z   (aplicando a cada fila)
# cada fila Z^(k) se transforma en X^(k)
X_mat = Z_mat @ L.T   # matriz 1000 × 4 vectorizado

# funcion color aleatorio
def color_random():
    return (random.random(), random.random(), random.random())

# histogramas de Z_i en una sola ventana con 4 subplots (2 filas x 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i in range(4):
    fila = i // 2     # 0 o 1
    col  = i % 2      # 0 o 1
    ax = axes[fila, col]
    
    ax.hist(Z_mat[:, i], bins=30, color=color_random())
    ax.set_title(fr"Histograma de $Z_{{{i+1}}} \sim N(0,1)$")
    ax.set_xlabel("Valor")
    ax.set_ylabel("Frecuencia")
plt.tight_layout()
plt.show()

# histogramas de X_i en una sola ventana con 4 subplots (2 filas x 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i in range(4):
    fila = i // 2     # 0 o 1
    col  = i % 2      # 0 o 1
    ax = axes[fila, col]

    ax.hist(X_mat[:, i], bins=30, color=color_random())
    ax.set_title(fr"Histograma de $W_{{{i+1}}}$ derivado de $LZ$")
    ax.set_xlabel("Valor")
    ax.set_ylabel("Frecuencia")
plt.tight_layout()
plt.show()
```

## Verificacion de la Covarianza

La covarianza estimada de los datos simulados es:

$$
\widehat{\text{Cov}}(X)
= \frac{1}{n - 1} X^\top X.
$$

Debe cumplirse que:

$$
\widehat{\text{Cov}}(X) \approx Cov
$$

La calidad de la aproximación se mide con la norma de Frobenius:

$$
\left\| \widehat{\text{Cov}}(X) - Cov \right\|_F
$$

# Modelo Geométrico Browniano Multivariado (GBM) y precio forward a 90 días

En este trabajo se modelan los precios de 4 activos (ALICORP, CREDICORP, FERREYCORP, SOUTHERN) usando un **Modelo Geométrico Browniano Multivariado (GBM)**, y se simula el precio forward a 90 días contado desde el último precio observado.

---

## 1. Datos de entrada del modelo

A partir de la serie histórica diaria de precios de cada activo, se construye:

1. Los **rendimientos diarios** de cada activo:

   - Si $x_{i,t}$ es el precio del activo $i$ en el día $t$, el rendimiento diario simple se define como  
     $$
     w_{i,t} = \frac{x_{i,t} - x_{i,t-1}}{x_{i,t-1}} 
     $$  
     o de forma equivalente  
     $$
     w_{i,t} = \frac{x_{i,t}}{x_{i,t-1}} - 1
     $$

2. A partir de estos rendimientos diarios se calculan:
   - $mean$ = vector de medias de rendimientos.
   - $cov$ = matriz de covarianza de los rendimientos.

3. También se considera:
   - $df\_final$ = DataFrame con la historia de precios.
   - El **último precio observado** como condición inicial del modelo: $x_0$.

---

## 2. Modelo continuo: GBM univariado

Para un solo activo:

$$
dX_t = \mu X_t\, dt + \sigma X_t\, dW_t.
$$

La solución exacta del GBM univariado es:

$$
X_t = X_0 \exp\left((\mu - \tfrac12\sigma^2)t + \sigma W_t\right).
$$

Esto implica que $X_t$ es log-normal y $\log(X_t)$ es normal.

---

## 3. Modelo GBM multivariado

Para $n$ activos se usa:

$$
dX_t = \mathrm{diag}(X_t)\left( \mu\, dt + L\, dW_t \right),
$$

donde:

- $\mu$ es un vector de medias de dimensión $n$.
- $W_t$ es un vector de Wiener estándar de dimensión $n$.
- $L$ es una matriz tal que $LL^\top = \Sigma$ (descomposición de Cholesky).
- $\Sigma$ es la matriz de covarianza de los rendimientos.

Esto permite reproducir correlaciones realistas entre activos.

---

## 4. Discretización diaria

Con paso temporal $\Delta t = dt$:

$$
X_{t+\Delta t} = X_t \odot \exp\left( (\mu - \tfrac12 \sigma^2)\Delta t + LZ\sqrt{\Delta t} \right),
$$

donde:

- $\sigma^2$ es el vector de varianzas ($\sigma_i^2 = \Sigma_{ii}$).
- $Z \sim N(0, I_n)$.
- $\odot$ indica multiplicación componente a componente.

---

## 5. Choques correlacionados

1. Se genera $Z \sim N(0,I_n)$.
2. Se aplica la matriz de Cholesky:

$$
\text{shock} = LZ\sqrt{\Delta t}.
$$

Esto asegura que:

$$
\text{shock} \sim N(0, \Sigma \Delta t).
$$

---

## 6. Fórmula de actualización discreta

El drift corregido es:

$$
drift_i = \left(\mu_i - \tfrac12\sigma_i^2\right)\Delta t.
$$

La actualización del GBM es:

$$
X_{t+1} = X_t \odot \exp(drift + shock).
$$

---

## 7. Precio forward a 90 días

El precio forward simulado a 90 días se obtiene iterando:

1. $X_0 = x_0$ (últimos precios observados).
2. Para $t = 1,\dots,90$:

   - Generar $Z_t \sim N(0,I_n)$.
   - Construir $shock_t = LZ_t\sqrt{dt}$.
   - Actualizar:

     $$
     X_t = X_{t-1} \odot \exp(drift + shock_t).
     $$

El vector final $X_{90}$ es el **precio forward simulado a 90 días**:

$$
X_{90} = (X_{90}^\text{ALICORP},\; X_{90}^\text{CREDICORP},\; X_{90}^\text{FERREYCORP},\; X_{90}^\text{SOUTHERN}).
$$

## SIMULACION DE PRECIO FORWRD t=90 dias
El modelo matemático del Movimiento Browniano Geométrico (GBM) multivariado aplicado a la simulación de precios forward a 90 días para cuatro activos financieros. Se deriva:

    * La dinámica estocástica del GBM multivariado.
    * La solución exacta del proceso.
    * La expresión del precio forward.
    * Las métricas estadísticas: media, varianza e intervalos de confianza.
    * La distribución marginal y conjunta del vector de precios simulados.



## Dinámica del GBM

Cada activo $X_i(t)$ sigue la ecuación estocástica:

\begin{equation*}
dX_i(t) = \mu_i X_i(t)\, dt + \sigma_i X_i(t)\, dW_i(t)
\end{equation*}

donde $W_i(t)$ es un proceso de Wiener correlacionado.

En forma vectorial:

\begin{equation*}
dX(t) = \mathrm{diag}(X(t))
\left( \mu\, dt + L\, dZ(t) \right)
\end{equation*}

donde:
- $Z(t) \sim N(0, I)$
- $L$ es la matriz de Cholesky tal que $LL^\top = \Sigma$


## Solución Exacta del GBM

\begin{equation*}
X(t) = X(0)\,
\exp\left( 
(\mu - \tfrac{1}{2}\sigma^2)t + LZ_t
\right),
\end{equation*}

con $Z_t \sim N(0, tI)$

En tiempo discreto diario con \(dt = 1/251\):

\begin{equation*}
X_{t+dt} =
X_t \exp\left(
(\mu - 0.5\sigma^2)dt + LZ\sqrt{dt}
\right)
\end{equation*}

## Precio Forward a 90 Días

Después de 90 pasos diarios:

\begin{equation*}
X_{90} = 
X_0
\prod_{k=1}^{90}
\exp\left(
(\mu - 0.5\sigma^2) dt + LZ_k\sqrt{dt}
\right)
\end{equation*}

Forma equivalente:

\begin{equation*}
X_{90} = X_0
\exp\left(
\sum_{k=1}^{90}
\left[
(\mu - 0.5\sigma^2) dt + LZ_k \sqrt{dt}
\right]
\right)
\end{equation*}


## Estadísticos Forward

### Media

\begin{equation*}
\bar{X}_{90} =
\frac{1}{N}
\sum_{k=1}^{N}
X_{90}^{(k)}.
\end{equation*}

### Varianza

\begin{equation*}
\mathrm{Var}(X_{90}) =
\frac{1}{N-1}
\sum_{k=1}^{N}
\left(
X_{90}^{(k)} - \bar{X}_{90}
\right)^2.
\end{equation*}

### Error estándar

\begin{equation*}
SE = 
\sqrt{
\frac{
\mathrm{Var}(X_{90})
}{N}
}.
\end{equation*}

### Intervalo de confianza del 95\%

\begin{equation*}
IC_{95\%} =
\left[
\bar{X}_{90} - 1.96\,SE,
\quad
\bar{X}_{90} + 1.96\,SE
\right].
\end{equation*}




## Distribución Marginal y Conjunta

Cada componente del vector $(X_{90})$ es aproximadamente lognormal:

\begin{equation*}
X_{90} \sim \mathrm{Lognormal}(\mu^*, \sigma^*),
\end{equation*}

con:

\begin{equation*}
\mu^* = (\mu - 0.5\sigma^2)T,
\qquad
\sigma^* = \sigma\sqrt{T}.
\end{equation*}


Distribución conjunta:

\begin{equation*}
\left(
X_{90}^{(1)},
X_{90}^{(2)},
X_{90}^{(3)},
X_{90}^{(4)}
\right),
\end{equation*}

preservada por:

\begin{equation*}
LZ_k \sim N(0, \Sigma dt).
\end{equation*}

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# PARÁMETROS
# -----------------------------
n = 4
dt = 1/251
T = 90
Nsim = 10000

cov = np.asarray(cov)
mu  = np.asarray(mean).astype(float)     # este será la media forward deseada
L   = np.linalg.cholesky(cov)

# Precio inicial
x0 = df_final[['ALICORP','CREDICORP','FERREYCORP','SOUTHERN']].iloc[-1].values.astype(float)

# -----------------------------
# DRIFT PARA FORZAR MEDIA FORWARD = μ
# -----------------------------
# Queremos: E[X_90] = mu
# X_90 = X0 * exp(90 * theta)  => theta = (1/90)*log(mu/x0)

theta = (np.log(mu / x0)) / T        # drift corregido

# Drift discreto para GBM
drift = (theta - 0.5 * np.diag(cov)) * dt

# -----------------------------
# MATRIZ PARA GUARDAR RESULTADOS
# -----------------------------
X_final = np.zeros((Nsim, n))

np.random.seed(123)

# -----------------------------
# SIMULACIONES MONTE CARLO
# -----------------------------
for k in range(Nsim):
    x = x0.copy()
    for t in range(1, T+1):
        z = np.random.randn(n)
        shock = L @ (z * np.sqrt(dt))
        x = x * np.exp(drift + shock)
    X_final[k, :] = x


# -----------------------------
# ESTADÍSTICAS
# -----------------------------
media_forward = X_final.mean(axis=0)
var_forward   = X_final.var(axis=0, ddof=1)

IC_low  = media_forward - 1.96 * np.sqrt(var_forward / Nsim)
IC_high = media_forward + 1.96 * np.sqrt(var_forward / Nsim)

activos = ['ALICORP','CREDICORP','FERREYCORP','SOUTHERN']

print("\nMEDIA forward a t=90 días (debe ser = μ):")
for i in range(n):
    print(f"{activos[i]:12s}: {media_forward[i]:.6f}")

print("\nμ (target deseado):")
for i in range(n):
    print(f"{activos[i]:12s}: {mu[i]:.6f}")

print("\nVARIANZA forward a t=90 días:")
for i in range(n):
    print(f"{activos[i]:12s}: {var_forward[i]:.6f}")

print("\nINTERVALOS DE CONFIANZA (95%):")
for i in range(n):
    print(f"{activos[i]:12s}: [{IC_low[i]:.6f}  ,  {IC_high[i]:.6f}]")


# ============================================================
#  HISTOGRAMAS INDIVIDUALES
# ============================================================
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i in range(n):
    axes[i].hist(X_final[:, i], bins=40)
    axes[i].set_title(f"Histograma Forward t=90 - {activos[i]}")

plt.tight_layout()
plt.show()

```
