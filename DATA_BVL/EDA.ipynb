### Analisis de la data de Alicorp S.A.A

```python
import pandas as pd

# Leer el CSV original separado por ;
df = pd.read_csv("ALICORP_SAA.csv", sep=';', header=None, engine='python')

# Extraer solo primera y última columna
data1 = df.iloc[:, [0, -1]].copy()
data1.columns = ['Fecha', 'Precio_Cierre']

# Convertir tipos de datos
data1['Fecha'] = pd.to_datetime(data1['Fecha'], format='%d/%m/%Y', errors='coerce')
data1['Precio_Cierre'] = pd.to_numeric(data1['Precio_Cierre'], errors='coerce')

# Eliminar filas donde la fecha o precio sean NaT/NaN
data1 = data1.dropna(subset=['Fecha', 'Precio_Cierre'])

#  Ordenar por fecha si lo deseas
data1 = data1.sort_values(by='Fecha').reset_index(drop=True)

print(data1.head(20))
```

```python
# Crear un rango completo de fechas
fecha_inicio = data1['Fecha'].min()
fecha_fin = data1['Fecha'].max()
rango_completo = pd.date_range(start=fecha_inicio, end=fecha_fin, freq='D')

# Fechas que faltan en el DataFrame
fechas_faltantes = rango_completo.difference(data1['Fecha'])

print("Fechas faltantes entre inicio y fin del conjunto de datos:")
print(fechas_faltantes)
```

```python
# ver si hay precios faltantes (NaN o vacíos) 
faltan_precios = data1['Precio_Cierre'].isna().sum()
print("Cantidad de precios faltantes (NaN):", faltan_precios)
```

### Unificacion de las datas y Ploteo de la informacion

```python
import pandas as pd
import matplotlib.pyplot as plt

# renombrar la columna 'Precio_Cierre' de cada DataFrame 
data1 = data1.rename(columns={'Precio_Cierre': 'ALICORP'})
data2 = data2.rename(columns={'Precio_Cierre': 'CREDICORP'})
data3 = data3.rename(columns={'Precio_Cierre': 'FERREYCORP'})
data4 = data4.rename(columns={'Precio_Cierre': 'SOUTHERN'})

# unir todas las tablas por la columna 'Fecha' 
df_final = data1.merge(data2, on='Fecha', how='outer') \
                .merge(data3, on='Fecha', how='outer') \
                .merge(data4, on='Fecha', how='outer')

# ordenar por fecha y limpiar índice 
df_final = df_final.sort_values(by='Fecha').reset_index(drop=True)

print(df_final.head())  # Ver que salió bien

# graficar precios 
plt.figure(figsize=(12, 6))
for columna in df_final.columns:
    if columna != 'Fecha':  # Evitar graficar la columna Fecha
        plt.plot(df_final['Fecha'], df_final[columna], label=columna)

plt.title("Variación de precios de acciones")
plt.xlabel("Fecha")
plt.ylabel("Precio de cierre")
plt.legend()
plt.grid()
plt.show()
```

Sea un portafolio compuesto por $n = 4$ activos financieros: ALICORP, CREDICORP, FERREYCORP y SOUTHERN. A partir de los precios diarios $X_{i,t}$, se definen los rendimientos simples como:
\begin{equation}
    w_{i,t} = \frac{X_{i,t} - X_{i,t-1}}{X_{i,t-1}} = \frac{X_{i,t}}{X_{i,t-1}} - 1,
\end{equation}
para cada activo $i = 1,2,3,4$ y cada día $t$.


Con estos rendimientos se construye la siguiente matriz:
\begin{equation}
R = 
\begin{pmatrix}
w_{1,1} & w_{2,1} & w_{3,1} & w_{4,1} \\
w_{1,2} & w_{2,2} & w_{3,2} & w_{4,2} \\
\vdots & \vdots & \vdots & \vdots \\
w_{1,T} & w_{2,T} & w_{3,T} & w_{4,T}
\end{pmatrix},
\end{equation}
donde cada columna corresponde a un activo y cada fila a un día.

La media de los rendimientos del activo $i$ se define como:
\begin{equation}
    \bar{w}_i = \frac{1}{T} \sum_{t=1}^{T} w_{i,t}.
\end{equation}

```python
import pandas as pd

df = pd.read_csv('DATA_UNIFICADA.csv')

# calcular rendimientos : (X_t - X_t-1) / X_t-1
rendimientos = df.copy()
rendimientos.set_index('Fecha', inplace=True)

for col in rendimientos.columns:
    rendimientos[col] = (rendimientos[col] - rendimientos[col].shift(1)) / rendimientos[col].shift(1)

# eliminar primera fila (NaN porque no tiene dia anterior)
rendimientos = rendimientos.dropna()

rendimientos.head()
```

La matriz de covarianza $\Sigma$ del sistema de activos se define como:
\begin{equation}
\Sigma =
\begin{pmatrix}
\text{Var}(w_1) & \text{Cov}(w_1,w_2) & \text{Cov}(w_1,w_3) & \text{Cov}(w_1,w_4) \\
\text{Cov}(w_2,w_1) & \text{Var}(w_2) & \text{Cov}(w_2,w_3) & \text{Cov}(w_2,w_4) \\
\text{Cov}(w_3,w_1) & \text{Cov}(w_3,w_2) & \text{Var}(w_3) & \text{Cov}(w_3,w_4) \\
\text{Cov}(w_4,w_1) & \text{Cov}(w_4,w_2) & \text{Cov}(w_4,w_3) & \text{Var}(w_4)
\end{pmatrix}.
\end{equation}



La covarianza entre los activos $i$ y $j$ mide cómo varían conjuntamente sus rendimientos:
\begin{equation}
\text{Cov}(w_i, w_j) = \frac{1}{T - 1} \sum_{t=1}^{T} (w_{i,t} - \bar{w}_i)(w_{j,t} - \bar{w}_j).
\end{equation}

```python
import pandas as pd
import numpy as np

# convertir a matriz numpy
R = rendimientos.values
T = len(R)  # numero de dias de datos
mean = np.mean(R, axis=0)  # promedio por activo

# matriz de covarianza manual 
cov = (R - mean).T @ (R - mean) / (T - 1)

# convertir a dataframe con nombres de columnas y filas
cov_df = pd.DataFrame(cov, 
        index=['ALICORP', 'CREDICORP', 'FERREYCORP', 'SOUTHERN'],
        columns=['ALICORP', 'CREDICORP', 'FERREYCORP', 'SOUTHERN'])

# impresion de la matriz de covarianza
cov_df
```
