## Análisis de Portafolio de Inversion
---
Se analizara el modelo 

$$f(c_i,x_i)=c_1.X_1+\cdot +c_n.X_n$$

donde: 
n: numero de activos financieros
X_i: precio de la accion i

---

Consideremos un portafolio formado por $n$ activos financieros. 
Denotamos por

* $X_i \quad (i = 1,\dots,n)$ el precio (o valor) del activo $i$ en el horizonte temporal de interés
* $c_i \quad (i = 1,\dots,n)$ el peso invertida en dicho activo.

El valor total del portafolio puede escribirse como la función
\begin{equation*}
f(\mathbf{c}, \mathbf{X}) 
= \sum_{i=1}^{n} c_i X_i,
\end{equation*}

donde
$
\mathbf{c} = (c_1,\dots,c_n)^{\mathsf{T}}, 
\,\,
\mathbf{X} = (X_1,\dots,X_n)^{\mathsf{T}}.
$

En forma vectorial, el valor del portafolio se escribe de manera compacta como
\begin{equation*}
f(\mathbf{c}, \mathbf{X}) = \mathbf{c}^{\mathsf{T}} \mathbf{X}.
\end{equation*}


Los pesos del portafolio satisfacen:
\begin{equation*}
\sum_{i=1}^{n} c_i = 1,
\qquad
c_i \in \mathbb{R} \quad (i=1,\dots,n),
\end{equation*}
y
\begin{equation*}
c_i \ge 0 \quad \text{para todo } i.
\end{equation*}


---

Sea un portafolio compuesto por $n = 4$ activos financieros: ALICORP, CREDICORP, FERREYCORP y SOUTHERN. A partir de los precios diarios $X_{i,t}$, se definen los rendimientos simples como:
\begin{equation*}
    w_{i,t} = \frac{X_{i,t} - X_{i,t-1}}{X_{i,t-1}} = \frac{X_{i,t}}{X_{i,t-1}} - 1,
\end{equation*}
para cada activo $i = 1,2,3,4$ y cada día $t$.


Con estos rendimientos se construye la siguiente matriz:
\begin{equation*}
R = 
\begin{pmatrix}
w_{1,1} & w_{2,1} & w_{3,1} & w_{4,1} \\
w_{1,2} & w_{2,2} & w_{3,2} & w_{4,2} \\
\vdots & \vdots & \vdots & \vdots \\
w_{1,T} & w_{2,T} & w_{3,T} & w_{4,T}
\end{pmatrix}
\end{equation*}
donde cada columna corresponde a un activo y cada fila a un día.

La media de los rendimientos del activo $i$ se define como:
\begin{equation*}
    \bar{w}_i = \frac{1}{T} \sum_{t=1}^{T} w_{i,t}
\end{equation*}

```python
import pandas as pd
import numpy as np

df = pd.read_csv('DATA_UNIFICADA.csv')

# calcular rendimientos : (X_t - X_t-1) / X_t-1
rendimientos = df.copy()
rendimientos.set_index('Fecha', inplace=True)

#for col in rendimientos.columns:
#    rendimientos[col] = (rendimientos[col] - rendimientos[col].shift(1)) / rendimientos[col].shift(1)

for col in rendimientos.columns:
    rendimientos[col] =np.log( (rendimientos[col] ) / rendimientos[col].shift(1))

# eliminar primera fila (NaN porque no tiene dia anterior)
rendimientos = rendimientos.dropna()

rendimientos.head()
```

La matriz de covarianza $\Sigma$ del sistema de activos se define como:
\begin{equation*}
Cov=\Sigma =
\begin{pmatrix}
\text{Var}(w_1) & \text{Cov}(w_1,w_2) & \text{Cov}(w_1,w_3) & \text{Cov}(w_1,w_4) \\
\text{Cov}(w_2,w_1) & \text{Var}(w_2) & \text{Cov}(w_2,w_3) & \text{Cov}(w_2,w_4) \\
\text{Cov}(w_3,w_1) & \text{Cov}(w_3,w_2) & \text{Var}(w_3) & \text{Cov}(w_3,w_4) \\
\text{Cov}(w_4,w_1) & \text{Cov}(w_4,w_2) & \text{Cov}(w_4,w_3) & \text{Var}(w_4)
\end{pmatrix}
\end{equation*}



La covarianza entre los activos $i$ y $j$ mide cómo varían conjuntamente sus rendimientos:
\begin{equation*}
\text{Cov}(w_i, w_j) = \frac{1}{T - 1} \sum_{t=1}^{T} (w_{i,t} - \bar{w}_i)(w_{j,t} - \bar{w}_j)
\end{equation*}

```python
import pandas as pd
import numpy as np

# convertir a matriz numpy
R = rendimientos.values
T = len(R)  # numero de dias de datos
mean = np.mean(R, axis=0)  # promedio por activo

# matriz de covarianza manual 
cov = (R - mean).T @ (R - mean) / (T - 1)

# convertir a dataframe con nombres de columnas y filas
cov_df = pd.DataFrame(cov, 
        index=['ALICORP', 'CREDICORP', 'FERREYCORP', 'SOUTHERN'],
        columns=['ALICORP', 'CREDICORP', 'FERREYCORP', 'SOUTHERN'])

# impresion de la matriz de covarianza
cov_df
```

Pasamos a corroborar si la data es consistente a trabes de la martiz de covarianza $Cov$, para ello debemos
comprobar:
* Si es simetrica
* Si es semidefinida positiva
* Se vera los autovalores de la matriz

Si se comprueba que matriz Cov es semidefinida positiva y simetrica(PSD), se hara la
descompisicion de Cholesky directamente.

En caso que la matriz $Cov$ arroje que no es seminidefinida positiva, la factorización de Cholesky falla, para corregirlo, se añade una pequeña cantidad positiva que se suma a la diagonal de una matriz para hacerla definida positiva. La pequeña cantidad se le denomina jitter y se calcula como:

\begin{equation*}
\text{jitter} 
= 
\max(0, -\lambda_{\min})
+
\varepsilon \, \frac{\operatorname{tr}(Cov)}{n},
\qquad
\varepsilon = 10^{-12}.
\end{equation*}

Se define entonces la matriz ajustada:
\begin{equation*}
Cov_j = Cov + \text{jitter} \, I.
\end{equation*}

Esta nueva matriz $Cov_j$, se garantiza que es semidefinida positiva (SPD), por lo que:
\begin{equation*}
Cov_j = L L^{\mathsf{T}}.
\end{equation*}

Para comprobar la exactitud numérica, se calcula el error de reconstrucción:
\begin{equation*}
\| Cov_j - L L^{\mathsf{T}} \|_F,
\end{equation*}

donde $\|\cdot\|_F$ denota la norma de Frobenius.  


Idealmente:
\begin{equation*}
\| Cov_j - L L^{\mathsf{T}} \|_F \rightarrow 0.
\end{equation*}

```python
import numpy as np
import pandas as pd


# hacemos una copia, con lo cual trabajaremos
cov = cov_df.to_numpy(copy=True)

# comprobaciones de simetría y autovalores

# simetria numerica
is_symmetric = np.allclose(cov, cov.T, atol=1e-12, rtol=1e-12)

# autovalores (para matrices simetricas usa eigvalsh)
eigvals = np.linalg.eigvalsh(cov)

# tolerancia numerica dependiente de la escala de la matriz
scale = max(1.0, np.linalg.norm(cov, ord=2))
tol = 1e-12 * scale

# es semidefinida positiva y simetrica(PSD) si todos los autovalores >= -tol
is_psd = np.all(eigvals >= -tol)

print("######   Comprobacion de la matriz Cov  para descomposicion ######")
print("La matriz Cov es simetrica:", is_symmetric)
print("Los autovalores son:", eigvals)
print("La matriz Cov es semidefinida positiva (numéricamente):", is_psd)

# descomposicion Cholesky
#   - Si es SPD (lambda_min > tol), Cholesky directo
#   - Si no, añadimos el jitter mínimo necesario para volverla SPD
lam_min = eigvals.min()
n = cov.shape[0]

def cholesky_with_min_jitter(A, lam_min, tol, safety=1e-12):
    """
    Intenta Cholesky directo. Si falla o A no es SPD,
    añade jitter = max(0, -lam_min) + safety * tr(A)/n y reintenta.
    """
    try:
        L = np.linalg.cholesky(A)
        jitter_used = 0.0
        return L, jitter_used
    except np.linalg.LinAlgError:
        pass

    # jitter minimo para empujar el menor autovalor por encima de 0
    jitter = max(0.0, -lam_min) + safety * (np.trace(A)/A.shape[0] if A.shape[0] > 0 else 1.0)
    A_j = A + jitter * np.eye(A.shape[0])
    L = np.linalg.cholesky(A_j)
    return L, jitter

L, jitter = cholesky_with_min_jitter(cov, lam_min, tol)
print("")
print("######   Descomposicion Cholesky de la matriz Cov   ######")
print(f"Valor jitter añadido para descomposicion Cholesky: {jitter:.3e}")

# descomposicion Cholesky
print("Matriz L de descomposicion Cholesky")
L_df = pd.DataFrame(L, index=cov_df.index, columns=[f"c{i+1}" for i in range(n)])
print("\nFactor de Cholesky L (triangular inferior):")
print(L_df)

# verificacion 
recon = L @ L.T
res_norm = np.linalg.norm((cov + jitter*np.eye(n)) - recon, ord='fro')
print("")
print(f"Verificacio de descomposicion: ||A(+jitter) - L L^T||_F = {res_norm:.3e}")
```

## Definición de la Variable Aleatoria Normal Estándar

Se considera un vector aleatorio
$$
Z = (Z_1, Z_2, Z_3, Z_4)^\top
$$
cuyos componentes son variables aleatorias independientes e idénticamente distribuidas:
$$
Z_i \stackrel{\text{i.i.d.}}{\sim} \mathcal{N}(0,1), \qquad i=1,\dots,4
$$

Al simular \(n = 1000\) valores de cada componente, obtenemos una matriz:

$$
Z_{\text{mat}} \in \mathbb{R}^{1000 \times 4},
$$

donde cada fila representa una realización de \(Z\).


---

## Construcción de la Variable Aleatoria Multivariada $X$

Se define:

$$
X = L Z
$$

y para cada una de las 1000 simulaciones:

$$
X^{(k)} = L Z^{(k)},
\qquad k=1,\dots,1000.
$$

De forma vectorizada:

$$
X = Z L
$$

Por teoría, se garantiza que:

$$
X \sim \mathcal{N}(0, \widetilde{\text{Cov}}).
$$

---

## Verificación Empírica

La covarianza estimada de los datos simulados es:

$$
\widehat{\text{Cov}}(X)
= \frac{1}{n - 1} X_{\text{mat}}^\top X_{\text{mat}}.
$$

Debe cumplirse que:

$$
\widehat{\text{Cov}}(X) \approx \widetilde{\text{Cov}}.
$$

La calidad de la aproximación se mide con la norma de Frobenius:

$$
\left\| \widehat{\text{Cov}}(X) - \widetilde{\text{Cov}} \right\|_F.
$$

---


```python
import numpy as np
import matplotlib.pyplot as plt
import random

# genera 4 variables aleatorias Z1,...,Z4 iid ~  N(0,1)
n = 1000 # cada una con 1000 valores
Z = [np.random.randn(n) for _ in range(4)]

# la matriz Z_mat esta ya ordenado para la multiplicacion
Z_mat = np.column_stack(Z)    # matriz 1000 × 5

# genera X = L Z   (aplicando a cada fila)
# cada fila Z^(k) se transforma en X^(k)
# vectorizado
X_mat = Z_mat @ L.T   # matriz 1000 × 4

# -------------------------------------------------------------------------
# Verificación: para un índice k, debe cumplirse X[k] = L @ Z[k]
k = 1
print("Z fila k:", Z_mat[k])
print("X fila k (vectorizado):", X_mat[k])
print("X fila k (producto directo):", L @ Z_mat[k].T)

# -------------------------------------------------------------------------
# funcion color aleatorio
def color_random():
    return (random.random(), random.random(), random.random())

# histogramas de Z_i en una sola ventana con 4 subplots (2 filas x 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i in range(4):
    fila = i // 2     # 0 o 1
    col  = i % 2      # 0 o 1
    ax = axes[fila, col]
    
    ax.hist(Z_mat[:, i], bins=30, color=color_random())
    ax.set_title(fr"Histograma de $Z_{{{i+1}}} \sim N(0,1)$")
    ax.set_xlabel("Valor")
    ax.set_ylabel("Frecuencia")
plt.tight_layout()
plt.show()

# histogramas de X_i en una sola ventana con 4 subplots (2 filas x 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i in range(4):
    fila = i // 2     # 0 o 1
    col  = i % 2      # 0 o 1
    ax = axes[fila, col]

    ax.hist(X_mat[:, i], bins=30, color=color_random())
    ax.set_title(fr"Histograma de $X_{{{i+1}}}$ derivado de $LZ$")
    ax.set_xlabel("Valor")
    ax.set_ylabel("Frecuencia")
plt.tight_layout()
plt.show()

# 7. Verificación de covarianza
cov_empirica = np.cov(X_mat, rowvar=False)

print("\nCovarianza empírica de X (aprox ≈ Cov original):")
print(cov_empirica)

print("\nCovarianza original:")
print(cov_df.to_numpy())

print("\nError Frobenius entre ambas (converge a 0 cuando n→∞):")
print(np.linalg.norm(cov_empirica - cov_df.to_numpy(), ord='fro'))


```

# Modelo Geométrico Log-Normal Multivariado (GBM)

```python
import numpy as np
import pandas as pd

# 'mean' : vector de medias de rendimientos (longitud n). Ya s calculo en el analisis anterior como `mean = np.mean(R, axis=0)`
# 'cov'  : matriz de covarianza (n x n)
# 'df_final' : DataFrame unificado con columnas ['Fecha','ALICORP','CREDICORP','FERREYCORP','SOUTHERN']

# parametros
n = 4                 # numero de activos del modelo
dt = 1/251            # paso diario de la data analizada

# aseguramos variables y tipos
cov = np.asarray(cov)                    # matriz (n x n)
mu = np.asarray(mean).astype(float)     # vector (n,)

# Precios iniciales, tomar la última fila de df_final (asegúrate que tenga las columnas correctas)
x0 = df_final[['ALICORP','CREDICORP','FERREYCORP','SOUTHERN']].iloc[-1].values.astype(float)

# reproducibilidad
np.random.seed(123)

# vector z ~ N(0, I) y choques correlacionados
z = np.random.randn(n)
shock = L @ (z * np.sqrt(dt))

# drift (nota: np.diag(cov) toma las varianzas)
drift = (mu - 0.5 * np.diag(cov)) * dt

# actualización GBM para un paso
x1 = x0 * np.exp(drift + shock)

print("Precios iniciales x0:", x0)
print("Precios simulados x1:", x1)
```
