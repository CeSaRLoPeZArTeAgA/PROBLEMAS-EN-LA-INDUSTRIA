```python
from scipy.optimize import linprog

# Min z = c^T x
c = [3, 2]  # coeficientes de la función objetivo
A = [[1, 2], [4, 1]]  # coeficientes de las restricciones
b = [8, 12]                # términos independientes
bounds = [(0, None), (0, None)]  # x1, x2 ≥ 0

res = linprog(c, A_ub=A, b_ub=b, method='highs')

print("Valor óptimo:", res.fun)
print("Variables:", res.x)
```

```python
import pulp

# Definir el problema de minimización
prob = pulp.LpProblem("Ejemplo_Simplex", pulp.LpMinimize)

# Variables
x1 = pulp.LpVariable("x1", lowBound=0)
x2 = pulp.LpVariable("x2", lowBound=0)

# Función objetivo
prob += 3*x1 + 2*x2

# Restricciones
prob += x1 + 2*x2 <= 8
prob += 4*x1 + x2 <= 12

# Resolver con CBC (solver fiable y ya incluido en PuLP)
prob.solve(pulp.PULP_CBC_CMD(msg=True))

# Resultados
print("Estado:", pulp.LpStatus[prob.status])
print("Valor óptimo:", pulp.value(prob.objective))
print("x1 =", pulp.value(x1))
print("x2 =", pulp.value(x2))
```

```python
from pyomo.environ import *

model = ConcreteModel()
model.x1 = Var(within=NonNegativeReals)
model.x2 = Var(within=NonNegativeReals)

# Función objetivo
model.obj = Objective(expr=3*model.x1 + 2*model.x2, sense=minimize)

# Restricciones
model.cons1 = Constraint(expr=model.x1 + 2*model.x2 <= 8)
model.cons2 = Constraint(expr=4*model.x1 <= 16)
model.cons3 = Constraint(expr=4*model.x2 <= 12)

# Resolver (por defecto usa GLPK si está instalado)
SolverFactory('glpk').solve(model)

print("Valor óptimo:", value(model.obj))
print("x1 =", value(model.x1))
print("x2 =", value(model.x2))
```
